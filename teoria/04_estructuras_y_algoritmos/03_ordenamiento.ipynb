## Divide y vencerás

- Complejidad de los algoritmos anteriores
    - La cantidad de pasos crece cuadráticamente.
    - Que el arreglo crezca en 1 puede incrementar sustancialmente la cantidad de pasos a realizar.
- Ordenar listas más pequeñas es sustancialmente más rápido que ordenar listas más grandes
- Ordenar una lista parcialmente ordenada es más rápido que una lista totalmente desordenada
- ¿Por qué no dividir el arreglo en partes más pequeñas y ordenar esas?
    - Esta estrategia es la que siguen los siguientes dos algoritmos que vamos a analizar, que alcanzan un mejor desempeño que los vistos anteriormente.
